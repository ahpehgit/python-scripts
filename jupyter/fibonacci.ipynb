```python
from functools import lru_cache

fibonacci_cache = {}

@lru_cache(maxsize=1000)
def fibonacci(n):
    if type(n) != int:
        raise TypeError("Not integer")
    elif n < 1:
        raise TypeError("Not positive integer")
    
    """
    # implement a cache function
    if n in fibonacci_cache:
        return fibonacci_cache[n]
    """
    
    if n == 1:
        return 1
    elif n == 2:
        return 1
    elif n > 2:
        value = fibonacci(n-1) + fibonacci(n-2)
        
    fibonacci_cache[n] = value
    return value

for i in range(1, 1101):
    print(i, ': ', fibonacci(i+1) / fibonacci(i)) # golden ratio: 1.61803
```
